```python
from PIL import Image
from PIL import ImageDraw
from ultralytics import YOLO
```

```python
# Load the YOLO model
model = YOLO("/content/drive/MyDrive/Traffic_Project/Kaggel/kaggel.pt", "v8")
```

```python
# Path to the image you want to process
source = "/content/drive/MyDrive/Traffic_Project/Results/image2.jpg"
```

```python
# Load the image
image = Image.open(source)
image
```

```python
# Perform predictions on the image
predictions = model(source)
```

```python
# Get the bounding boxes along with their confidence scores
boxes = predictions[0].boxes.xyxy
```

```python
# Print the total number of vehicles detected
print("Total number of Vehicles:", len(boxes))
```

```python
# Draw bounding boxes on the image
draw = ImageDraw.Draw(image)
for box in boxes:
    draw.rectangle(box.tolist(), outline="red")
```
